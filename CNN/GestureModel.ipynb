```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Bidirectional, Dense, Dropout, LSTM, GRU
import numpy as np
import os
```

```python
epoch = 20  # epoch数
# データの作成
# データ用の変数
input_data = []
label_data = []
id = 0
data_dir = './data'#dataフォルダの下にデータがある
for c in sorted(os.listdir(data_dir)):#フォルダを取り出す
  print(f'folder: {c}, label: {id}')#フォルダごとにラベルを付ける
  d = os.path.join(data_dir, c)
  files = os.listdir(d)
  for i in [ft for ft in files if ('txt' in ft)]:#txtファイルのみ取り出す
    with open(os.path.join(d, i), mode='r') as f:
      temp = []
      for line in f.readlines():#データの読み込み
        t = line.strip().split(',')
        temp.append(t)
      temp = np.array(temp, dtype=np.float32)
      ave = np.mean(temp)
      var = np.std(temp)
      temp = (temp - ave) / var  # 正規化
      input_data.append(temp)
      label_data.append(id)
  id += 1
train_data = np.array(input_data, dtype=np.float32)#学習データ（入力）
train_label = np.array(label_data)#学習データ（ラベル）
```

```python
model = Sequential()
model.add(LSTM(128, input_shape=(50, 3)))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(4, activation='softmax'))
```

```python
#    model = keras.models.load_model(os.path.join('result', 'RNNmodel'))  # modelのロード
#モデルの設定
#model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# TensorBoard用の設定
tb_cb = keras.callbacks.TensorBoard(log_dir='log', histogram_freq=1)
#学習の実行
model.fit(train_data, train_label, epochs=epoch, batch_size=16, callbacks=[tb_cb])
```

```python
model.save_weights('gesture_model.h5')
```
